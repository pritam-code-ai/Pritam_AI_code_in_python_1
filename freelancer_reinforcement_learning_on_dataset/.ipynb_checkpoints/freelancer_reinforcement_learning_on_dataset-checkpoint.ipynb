```python
def MCModelv1(data, alpha, e, epsilon, budget, reward):
    # Define the States
    number_of_programming_language_used = list(set(data['number_of_programming_language_used']))
    # Initialise V_0
    V0 = data['V_0']
    data['V'] = V0
    output = []
    output1 = []
    output2 = []
    actioninfull = []
    #Interate over the number of episodes specified
    for e in range(0,e):
        print("\n\nfor e in range(0,e)")
        print("e at : {}".format(e))
        print("-------------------------------------------------------------------- \n\n\n")
        episode_run = []
        #Introduce epsilon-greedy selection, we randomly select the first episode as V_0(a) = 0 for all actions
        epsilon = epsilon        
        if e == 0:
            print("if e == 0")
            print("for i in range(0,len(number_of_programming_language_used)) : ")
            print("episode_run = np.append(episode_run,np.random.random_integers(low = 1, high = sum(1 for p in data.iloc[:, 0] if p == i+1 ), size = None)) : \n")
            for i in range(0,len(number_of_programming_language_used)):
                episode_run = np.append(episode_run,np.random.random_integers(low = 1, high = sum(1 for p in data.iloc[:, 0] if p == i+1 ), size = None))
                print("episode_run = {} \n\n\n".format(episode_run))
            episode_run = episode_run.astype(int)
            print("episode_run = episode_run.astype(int) : {}\n\n\n".format(episode_run))
            
        else:
            print("not e == 0 or e != 0 or else: ")
            print("for i in range(0,len(number_of_programming_language_used): \n\n")
            for i in range(0,len(number_of_programming_language_used)):
                print("greedyselection = np.random.random_integers(low = 1, high =10) : ")      
                greedyselection = np.random.random_integers(low = 1, high =10)
                print("greedyselection = {} \n".format(greedyselection))
               
                
                if greedyselection <= (epsilon)*10:
                    print("if greedyselection <= (epsilon)*10:")
                    print("episode_run = np.append(episode_run,np.random.random_integers(low = 1, high = sum(1 for p in data.iloc[:, 0] if p == i+1 ), size = None)) : \n")  
                    episode_run = np.append(episode_run,np.random.random_integers(low = 1, high = sum(1 for p in data.iloc[:, 0] if p == i+1 ), size = None))
                    print("episode_run = {} \n\n\n".format(episode_run))
                else:
                  
                    print("greedyselection > (epsilon)*10 or else: ")
                    print("for i in range(0,len(number_of_programming_language_used)): \n")
                    print("data_I = data[data['number_of_programming_language_used'] == (i+1)]  \n")
                    data_I = data[data['number_of_programming_language_used'] == (i+1)] 
                    print("data_I {}\n".format(data_I))
                    MaxofVforI = data_I[data_I['V'] == data_I['V'].max() ]['Software']
                    print("MaxofVforI = data_I[data_I['V'] == data_I['V'].max() ]['Software'] : ")
                    print("MaxofVforI = {} \n".format(MaxofVforI))
                    #If multiple max values, take first
                    print("If multiple max values, take first, MaxofVforI = MaxofVforI.values[0] : ")
                    MaxofVforI = MaxofVforI.values[0]
                    print("MaxofVforI = {} \n".format(MaxofVforI))
                    episode_run = np.append(episode_run, MaxofVforI)
                    print("episode_run = np.append(episode_run, MaxofVforI) : ")
                    print("episode_run = {} \n\n\n".format(episode_run))
                  
                episode_run = episode_run.astype(int)
                print("episode_run = episode_run.astype(int) : ")
                print("episode_run = {} \n".format(episode_run))
                
                           
            episode = pd.DataFrame({'number_of_programming_language_used' : number_of_programming_language_used, 'Software': episode_run})    
            episode['Merged_label'] =  (episode['number_of_programming_language_used']*10 + episode['Software']).astype(float)
            data['new_Merged_label'] = (data['new_Merged_label']).astype(float)
            data['Reward'] = reward
            episode2 =  episode.merge(data[['new_Merged_label','Actual_price','Reward']], left_on='Merged_label',right_on='new_Merged_label', how = 'inner')
            data = data.drop('Reward',1)
        
            # Calculate our terminal reward
            if(budget >= episode2['Actual_price'].sum()):
                Return = 1  
            else:
                Return = -1 
            episode2 = episode2.drop('Reward',1)
            episode2['Return'] = Return
        
            # Apply update rule to actions that were involved in obtaining terminal reward 
            data = data.merge(episode2[['Merged_label','Return']], left_on='new_Merged_label',right_on='Merged_label', how = 'outer')
            data['Return'] = data['Return'].fillna(0)
            for v in range(0,len(data)):
                if data.iloc[v,7] == 0:
                    data.iloc[v,5] = data.iloc[v,5] 
            else:
                data.iloc[v,5]  = data.iloc[v,5]  + alpha*( (data.iloc[v,7]/len(number_of_programming_language_used)) - data.iloc[v,5] )
                
            # Output table    
            data = data.drop('Merged_label',1)
            data = data.drop('Return',1)
            # Output is the Sum of V(a) for all episodes
        output  = np.append(output, data.iloc[:,-1].sum())
        
        # Output 1 and 2 are the Sum of V(a) for for the cheapest actions and rest respectively
        # I did this so we can copare how they converge whilst applying to such a small sample problem
        output1 = np.append(output1, data.iloc[[1,2,4,8],-1].sum())
        output2 = np.append(output2, data.iloc[[0,3,5,6,7],-1].sum())
        
        # Ouput to optimal action from the model based on highest V(a)
        action = pd.DataFrame(data.groupby('number_of_programming_language_used')['V'].max())
        action2 = action.merge(data, left_on = 'V',right_on = 'V', how = 'inner')
        action3 = action2[['number_of_programming_language_used','Software']]
        action3 = action3.groupby('number_of_programming_language_used')['Software'].apply(lambda x :x.iloc[np.random.randint(0, len(x))])
        
        # Output the optimal action at each episode so we can see how this changes over time
        actioninfull = np.append(actioninfull, action3)
        actioninfull = actioninfull.astype(int)
        
        # Rename for clarity
        SumofV = output
        SumofVForCheapest = output1
        SumofVForExpensive = output2
        OptimalActions = action3
        ActionsSelectedinTime = actioninfull
        
    return(SumofV, SumofVForCheapest, SumofVForExpensive, OptimalActions, data, ActionsSelectedinTime)

```

```python
alpha = 0.1
num_episodes = 100
epsilon = 0.5
budget = 30

# Currently not using a reward
reward = [0,0,0,0,0,0,0,0,0]

start_time = time.time()

Mdl = MCModelv1(data=data, alpha = alpha, e = num_episodes,epsilon = epsilon, budget = budget, reward = reward)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
number_of_programming_language_used = list(set(data['number_of_programming_language_used']))
actions = pd.DataFrame()

for a in range(0, len(number_of_programming_language_used)):   
    individualactions = []
    for i in range(0,num_episodes):    
        individualactions = np.append(individualactions, Mdl[5][a+(i*(len(number_of_programming_language_used)))])
    actions[a] = individualactions
    plt.plot(range(0,num_episodes), actions[a])
    plt.title('Software for number_of_programming_language_used: ' + str(a+1))
    plt.xlabel('Episode')
    plt.ylabel('Software')
    plt.show()
```

```python
actions2 = actions
actions2['Software1'] = actions2.iloc[:,0]+10
actions2['Software2'] = actions2.iloc[:,1]+20
actions2['Software3'] = actions2.iloc[:,2]+30
actions2['Software4'] = actions2.iloc[:,3]+40

actions3 = actions2.merge(data[['new_Merged_label','Actual_price']],left_on = 'Software1',right_on = 'new_Merged_label', how = 'left')
actions4 = actions3.merge(data[['new_Merged_label','Actual_price']],left_on = 'Software2',right_on = 'new_Merged_label', how = 'left')
actions5 = actions4.merge(data[['new_Merged_label','Actual_price']],left_on = 'Software3',right_on = 'new_Merged_label', how = 'left')
actions6 = actions5.merge(data[['new_Merged_label','Actual_price']],left_on = 'Software4',right_on = 'new_Merged_label', how = 'left')


actions6['Total_Cost'] = actions6.iloc[:,9] + actions6.iloc[:,11] + actions6.iloc[:,13] + actions6.iloc[:,15]
actions6 = actions6.iloc[:,[0,1,2,3,-1]]

actions6 = actions6.iloc[:num_episodes]

plt.plot(range(0,num_episodes), actions6['Total_Cost'])
plt.plot([0, num_episodes], [budget, budget], 'k-', lw=2)
plt.title('Total actual price of Best Softwares at each Episode')
plt.xlabel('Episode')
plt.ylabel('Total actual price ()')
plt.ylim([0,budget+10])
plt.show()
```
